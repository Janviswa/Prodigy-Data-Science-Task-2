```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\janani\OneDrive\Documents\Titanic\train.csv" 
titanic_df = pd.read_csv(file_path)
```

```python
titanic_df
```

```python
print(titanic_df.head())

print(titanic_df.describe())

print(titanic_df.info())

print(titanic_df.isnull().sum())

titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

titanic_df.drop(columns=['Cabin'], inplace=True)

titanic_df.dropna(subset=['Fare'], inplace=True)

print(titanic_df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
titanic_df['Age'].hist(bins=30, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
titanic_df['Fare'].hist(bins=30, edgecolor='black')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=titanic_df, palette='viridis')
plt.title('Count of Passengers by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=titanic_df, palette='viridis')
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', data=titanic_df, palette='viridis')
plt.title('Count of Passengers by Embarked')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.show()
```
